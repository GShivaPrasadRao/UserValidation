```python
q = int(input("Enter R2 Value"))
```

```python
pn = int(input("Enter Prime Number"))
```

```python
th = int(input("Enter Threshold Value"))
```

```python
from numpy import sqrt 
puk=p*q+pn
ip=(pn-p)*(q-p)+th
ipr=ip*sqrt(puk)-th
ik=ipr-ip+pow(q,p)
ik=int(ik+cc+max(ip,ipr))
ik=ik & pn
ik=ik+pn-(q-p)
pubkey=ik<<pn
pubk=pubkey>>(q-p)
```

```python
pubkeyL=[]
s=0
for i in ton:
    puk=p*q+pn
    ip=(pn-p)*(q-p)+th
    ipr=ip*sqrt(puk)-th
    ik=ipr-ip+pow(q,p)
    ik=int(ik+cc+max(ip,ipr))
    ik=ik & pn
    ik=ik+pn-(q-p)
    pubkey=ik<<pn
    pubk=pubkey>>(q-p)
    pubk=pubk++ton[s]
    s=s+1
    pubkeyL.append(pubk)
```

```python
p = int(input("Enter p Value")) 
q = int(input("Enter q Value")) 
r = int(input("Enter r Value")) 
s = int(input("Enter s Value")) 
```

```python
prikeyL=[]
w=0
for i in ton:
    pri=p & s
    ipk=q | r
    k1=int(ipk/pri)
    tt = datetime.now()
    cc=int(tt.strftime("%H%M%S"))
    ma=max(p,s)
    k2=pubkeyL[w]+cc+max(k1,ma)+ton[w]
    k3=(k1<<k2)
    k3=k3<<q
    k4=r>>k3
    k4=k4 & (p+s)
    ikpr1=(k1 & k3)+(k2 | k4)
    ikpr2=ikpr1+(min(k1,k3)^ max(k2,k4))+pubkeyL[w]
    prike=(ikpr2<<pri)>>ipk+ton[w]
    ip=int(ip)
    ipr=int(ipr)
    priks=ip^ipr
    prke=prike ^ priks+ton[w]
    w=w+1
    prikeyL.append(prke)
```

```python
M=len(ton)
f=0
print("Group Key Pairs Generated are")
for i in ton:
  print(pubkeyL[f],":",prikeyL[f])
  f=f+1
```
